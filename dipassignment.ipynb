```python
#Program to obtain negative of an image 
```

```python
def negative(im):
    height=len(im)
    width = len(im[0])
    for row in range(height):
        for col in range(width):
            red = 255 - im[row][col][0]
            green = 255 - im[row][col][1]
            blue = 255 - im[row][col][2]
            im[row][col]=[red,green,blue]
    return im
negim = cv2.imread('DIPC3/negative.tif')
img = cv2.cvtColor(negim, cv2.COLOR_BGR2RGB)
negimage = negative(negim)
nimg = cv2.cvtColor(negimage, cv2.COLOR_BGR2RGB)
plt.subplot(121),plt.imshow(img,cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(nimg,cmap='gray'),plt.title('Negative Image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
def n(im):
    g = 1 * (np.log(1 + np.float32(im)))
    return g

bimg = cv2.imread('DIPC3/nolog.tif',0)
plt.subplot(121),plt.imshow(bimg,cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(n(bimg),cmap='gray'),plt.title('Log Transformed Image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
def p(im):
    g = 1.8*np.power(im,2.7)
    return g

bimg = cv2.imread('DIPC3/washedout.tif',0)
plt.subplot(121),plt.imshow(bimg,cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(p(bimg),cmap='gray'),plt.title('Power Transform Image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv2.imread('DIPC4/Blown.tif',0)

img_float32 = np.float32(img)

dft = cv2.dft(img_float32, flags = cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)

rows, cols = img.shape
crow, ccol = rows/2 , cols/2     # center

mask = np.zeros((rows, cols, 2), np.uint8)
mask[crow-20:crow+20, ccol-20:ccol+20] = 1

# apply mask and inverse DFT
fshift = dft_shift*mask
f_ishift = np.fft.ifftshift(fshift)
img_back = cv2.idft(f_ishift)
img_back = cv2.magnitude(img_back[:,:,0],img_back[:,:,1])

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img_back, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])

plt.show()                
```

```python
from skimage.io import imread
import matplotlib.pyplot as plt
import scipy.fftpack as fp

im = np.mean(imread('lenad.png'), axis=2) # assuming an RGB image
plt.figure(figsize=(5,5))
plt.imshow(im, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
F1 = fp.fft2((im).astype(float))
F2 = fp.fftshift(F1)
plt.figure(figsize=(5,5))
plt.imshow( (20*np.log10( 0.1 + F2)).astype(int), cmap=plt.cm.gray)
plt.show()

(w, h) = im.shape
half_w, half_h = int(w/2), int(h/2)

# high pass filter
n = 25
F2[half_w-n:half_w+n+1,half_h-n:half_h+n+1] = 0 # select all but the first 50x50 (low) frequencies
plt.figure(figsize=(5,5))
plt.imshow( (20*np.log10( 0.1 + F2)).astype(int),cmap=plt.cm.gray)
plt.show()

im1 = fp.ifft2(fp.ifftshift(F2)).real
plt.figure(figsize=(5,5))
plt.imshow(im1, cmap='gray')
plt.axis('off')
plt.show()
```
